```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils

y_age = df[2]
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(y_age)
encoded_Y = encoder.transform(y_age)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y_age = np_utils.to_categorical(encoded_Y)
print(dummy_y_age)
```

```python
N = df.shape[0]
print(N)
imgs = [None] * N

from sklearn.utils import class_weight

y_ints = [y.argmax() for y in dummy_y_age]



class_weights = class_weight.compute_class_weight('balanced',
                                                 np.unique(y_ints),
                                                 y_ints)

class_weight_dict = dict(enumerate(class_weights))
class_weight_dict
```

```python
for i in range(N):
    I = Image.open(df[0][i])
    imgs[i] = np.array(I)
    imgs[i]=np.stack(imgs[i],axis=0) / (255.0)
    c=imgs[i].shape
    print(c)
print(imgs[0])
```

```python
from sklearn.model_selection import KFold

kfold = KFold(n_splits = 5, shuffle = True, random_state=10)
```

```python
idg = ImageDataGenerator(width_shift_range=0.1,
                         height_shift_range=0.1,
                         zoom_range=0.2,
                         fill_mode='nearest',
                         horizontal_flip= True, vertical_flip= True)
```

```python
VALIDATION_ACCURACY = []
VALIDATION_LOSS = []
save_dir = 'C:/Users/Mike/saved_models_trial4/'
fold_var = 1


def get_model_name(k):
    return 'model_'+str(k)+'.h5'


for train_index, val_index in kfold.split(imgs,dummy_y_age):
    training_data = df.iloc[train_index]
    validation_data = df.iloc[val_index]
    
    train_data_generator = idg.flow_from_dataframe(training_data,x_col = 0, y_col = 2 ,
                                                   class_mode = "sparse", shuffle = True, target_size=(314,640))
    valid_data_generator  = idg.flow_from_dataframe(validation_data, x_col = 0, y_col = 2,
                                                    class_mode = "sparse", shuffle = True, target_size=(314,640))
    
    # create model
    input_layer = Input(shape=(314,640,3));
    
    x = Conv2D(filters=16,kernel_size=(3,3),activation='relu')(input_layer)
    x = Conv2D(filters=16,kernel_size=(3,3),activation='relu')(x)
    x = MaxPooling2D(pool_size=(2,2), strides=(2,2))(x)
    
    
    x = Conv2D(filters=32,kernel_size=(3,3),activation='relu')(x)
    x = MaxPooling2D(pool_size=(2,2), strides=(2,2))(x)
    
    x = Conv2D(filters=64,kernel_size=(3,3),activation='relu')(x)
    x = Conv2D(filters=64,kernel_size=(3,3),activation='relu')(x)
    x = MaxPooling2D(pool_size=(2,2),strides=(2,2))(x)
                     
    x = Conv2D(filters=128,kernel_size=(3,3),activation='relu')(x)
    x = Conv2D(filters=128,kernel_size=(3,3),activation='relu')(x)
    x = MaxPooling2D(pool_size=(2,2),strides=(2,2))(x)
    
    x = Conv2D(filters=256,kernel_size=(3,3),activation='relu')(x)
    x = Conv2D(filters=256,kernel_size=(3,3),activation='relu')(x)
    x = MaxPooling2D(pool_size=(2,2),strides=(2,2))(x)
    
    x = Conv2D(filters=512,kernel_size=(3,3),activation='relu')(x)
    x = Conv2D(filters=1024,kernel_size=(3,3),activation='relu')(x)
    x = MaxPooling2D(pool_size=(2,2),strides=(2,2))(x)
    
    
    x = Flatten()(x)
    
    x = Dense(32,activation='relu')(x)
    x = Dropout(0.2)(x)
    x = Dense(3,activation='softmax')(x)
    
    model = Model(inputs=[input_layer],outputs=[x])
    model.summary()
    
    checkpoint = keras.callbacks.ModelCheckpoint(save_dir+get_model_name(fold_var), 
                                                    monitor='val_accuracy', verbose=1, 
                                                    save_best_only=True, mode='max')
    
    callbacks_list = [checkpoint]
    
    # compile model
    model.compile(loss='sparse_categorical_crossentropy', optimizer=SGD(lr=0.001, decay=0.09,momentum=0.9,
                                                                        nesterov=True), metrics=['accuracy'])
    
    history = model.fit(train_data_generator, epochs=7,callbacks=callbacks_list, 
                        validation_data=valid_data_generator, class_weight = class_weights)

    # load best model to evaluate the performance of the model
    
    model.load_weights("C:/Users/Mike/saved_models_trial4/model_"+str(fold_var)+".h5")
    
    results = model.evaluate(valid_data_generator)
    results = dict(zip(model.metrics_names,results))

    VALIDATION_ACCURACY.append(results['accuracy'])
    VALIDATION_LOSS.append(results['loss'])

    tf.keras.backend.clear_session()
    
    fold_var += 1
```
